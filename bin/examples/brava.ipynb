```python
import numpy as np
import pylab as plt


from netCDF4 import Dataset as NF

it=-1

ifid=NF('snap3D.cdf',mode='r')

yt = ifid.variables['yt'][:]
zt = ifid.variables['zt'][:]
U = ifid.variables['U'][:]
Nsqr = ifid.variables['Nsqr'][:]

E_int0 = ifid.variables['E'][0,:]
E_int = ifid.variables['E'][it,:]
meanfl_int = ifid.variables['meanfl'][it,:]

diss_ks   =  ifid.variables['diss_ks'][it,:]
diss_ke   =  ifid.variables['diss_ke'][it,:]
diss_ls   =  ifid.variables['diss_ls'][it,:]
diss_le   =  ifid.variables['diss_le'][it,:]
diss_ms   =  ifid.variables['diss_ms'][it,:]
diss_me   =  ifid.variables['diss_me'][it,:]
ifid.close()


diss = diss_ke-diss_ks
diss += diss_le-diss_ls
diss += diss_me-diss_ms

x=0

plt.figure(figsize=(16,9))


ax = plt.subplot(3,3,1)
co = plt.contourf(yt/1e3,zt, U[:,:,x]   ) 
plt.colorbar(co)

ax = plt.subplot(3,3,2)
ax.plot( Nsqr[:,1,x]**0.5/1e-4,zt)

ax = plt.subplot(3,3,3)
co = plt.contourf(yt/1e3,zt, E_int0[:,:,x]*1e3   ) 
plt.colorbar(co)

ax = plt.subplot(3,3,4)
co = plt.contourf(yt/1e3,zt, E_int[:,:,x]   ) 
plt.colorbar(co)

ax = plt.subplot(3,3,5)
co = plt.contourf(yt/1e3,zt, meanfl_int[:,:,x]  ) 
plt.colorbar(co)

ax = plt.subplot(3,3,6)
co = plt.contourf(yt/1e3,zt, diss[:,:,x] /Nsqr[:,:,x]*0.2/(1+0.2) *1e4  ) 
plt.colorbar(co)

plt.show()
```
